```python
def GridSearch_BestParam(x, y, clf, param_grid,cv=10):
    grid_search = GridSearchCV(clf,
                              param_grid=param_grid,
                              cv=cv)
    start= time()
    grid_search.fit(x,y)
    top_params=grid_search.best_params_
    return top_params
```
